## Data Analysis

Questions:

```python
#convert from $ to million $
Movies['budget']= Movies['budget'] /1000000
Movies['revenue']= Movies['revenue'] /1000000
```

### Find Movies with the highest profit

```python
#Calculating profit
Movies['profit']=Movies['revenue'] - Movies ['budget']
Movies
```

```python
#Sorting movies by profit
Movies.sort_values(by='profit', ascending = False)
```

### Popularity 

```python
#Top 10 movies
Top_10 = Movies.sort_values(by='profit', ascending = False).head(10)
Top_10
```

### Best Directors

```python
#Top 10 directors
Top10directors=Movies.groupby ('director').popularity.mean().sort_values(ascending=False).head (5)
Top10directors
```

```python

Top10directors.plot(x ='Director', y='Popularity', kind = 'bar')	
Top10directors.plot.bar(title="Diector Popularity");
plt.xlabel("Director")
plt.ylabel("Popularity")
plt.xticks(rotation=45)
plt.show()
```

This shows that of the top 5 directors, Colin Trevorrow was the most popular while Don Hall and Chris williams were the least popular.

```python
plt.boxplot(Movies['revenue'])
plt.show()
```

This shows that the revenue has a lot of outliers, because the zero values weren't dropped

### Popular Genres

```python
#Top 10 popular genres
Movies.head(3)
```

```python
#Spliting multiple genres into separate columns
genres_Movies = Movies['genres'].str.split("|", expand=True)
genres_Movies.head(5)
```

```python
# Creating a separate dataframe form unique genres records.
genres_Movies = genres_Movies.stack()

genres_Movies = pd.DataFrame(genres_Movies)
genres_Movies.head()
```

```python
#Idendifying the most popular genres using value_counts
genres_Movies.rename(columns={0:'genres_adj'}, inplace=True)
count_genres = genres_Movies.genres_adj.value_counts()
count_genres
```

```python
plt.figure(figsize = [20,10]) 
plt.bar(count_genres.index, count_genres)
plt.xticks(count_genres.index, rotation = 60)
plt.xlabel("Genres")
plt.ylabel("Popularity")
plt.title("Most Popular Genres")
```

The above barchart shows that the most popular genres are Drama followed by Comedy then Action.
